```python
from datasets import load_dataset, Features, Value
import pandas as pd

features = Features({
    "chunk_id": Value("string"),
    "chunk_text": Value("string"),
    "original_text_id": Value("string"),
    "original_text": Value("string"),
    "position": Value("int64"),
    "num_chunks_in_doc": Value("int64"),
    "num_tokens": Value("int64"),
    "msa_translation": Value("string"),
    "cleaned_msa_translation": Value("string"),
    "semantic_similarity": Value("float32"),
    "lm_fluency_score": Value("float32"),
    "composite_score": Value("float32"),
    "quality_flag": Value("bool"),
    "norm_fluency": Value("float32")  # Add the missing column
})

raw_dataset = load_dataset(
    "tunis-ai/tunisian-msa-parallel-corpus-evaluated",
    split="train",
    features=features
)

df = pd.DataFrame(raw_dataset)
print(df.head())
print(df.info())
```

```python
import numpy as np
from sentence_transformers import SentenceTransformer, util
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
import math
from tqdm import tqdm

# --- Load Models ---
print("Loading models...")
# Use LaBSE for semantic similarity (multilingual, good for Arabic varieties)
semantic_model = SentenceTransformer('sentence-transformers/LaBSE')
# Use AraGPT2 for fluency scoring
fluency_tokenizer = AutoTokenizer.from_pretrained('aubmindlab/aragpt2-base')
fluency_model = AutoModelForCausalLM.from_pretrained('aubmindlab/aragpt2-base')
fluency_model.eval() # Set to evaluation mode

# --- Prepare Text Lists ---
# Use the cleaned MSA translation for evaluation
tunisian_texts = df['chunk_text'].astype(str).tolist()
msa_texts_eval = df['cleaned_msa_translation'].astype(str).tolist()

# --- 1. Calculate LaBSE Semantic Similarity ---
print("Calculating LaBSE similarity...")
emb_t = semantic_model.encode(tunisian_texts, convert_to_tensor=True, show_progress_bar=True)
emb_m = semantic_model.encode(msa_texts_eval, convert_to_tensor=True, show_progress_bar=True)
cos_sim = util.cos_sim(emb_t, emb_m).diagonal().cpu().numpy()

# --- 2. Calculate AraGPT2 Perplexity (Fluency) ---
print("Calculating AraGPT2 perplexity...")
def calculate_perplexity(text, model, tokenizer):
    enc = tokenizer(text, return_tensors="pt", truncation=True, max_length=512)
    input_ids = enc.input_ids.to(model.device)
    if input_ids.shape[1] == 0: # Check for empty tokenization
        return float('inf') # High perplexity for empty/invalid input
    with torch.no_grad():
        outputs = model(input_ids, labels=input_ids)
        loss = outputs.loss.item()
    return math.exp(loss)

ppls = []
for msa_text in tqdm(msa_texts_eval, desc="Perplexity"):
    ppl = calculate_perplexity(msa_text, fluency_model, fluency_tokenizer)
    ppls.append(ppl)
ppls = np.array(ppls)

# --- 3. Calculate Composite Score ---
print("Calculating composite score...")
# Normalize LaBSE scores to [0,1]
cos_norm = (cos_sim - cos_sim.min())/(cos_sim.max()-cos_sim.min()+1e-12)
# Normalize perplexity scores: lower perplexity is better, so invert after log normalization
log_ppls = np.log(ppls)
ppl_norm = (log_ppls - log_ppls.min())/(log_ppls.max()-log_ppls.min()+1e-12)
ppl_score = 1 - ppl_norm # Invert so higher score means lower perplexity (better fluency)

# Combine scores (adjust weights as needed)
W_SEMANTIC = 0.7 # Weight for semantic similarity
W_FLUENCY = 0.3  # Weight for fluency
composite = W_SEMANTIC * cos_norm + W_FLUENCY * ppl_score

# Add scores back to the DataFrame
df['labse_similarity'] = cos_sim
df['aragpt2_perplexity'] = ppls
df['norm_labse_similarity'] = cos_norm
df['norm_fluency_score'] = ppl_score
df['composite_score_v2'] = composite

# --- 4. Create Quality Splits based on Composite Score ---
print("Creating quality splits...")
# Sort by composite score (descending)
idx_sorted = np.argsort(-composite) # Descending order
N = len(composite)
# Define thresholds (adjust percentages as needed)
GOLD_PERCENT = 0.10
SILVER_PERCENT = 0.30

gold_idx = idx_sorted[:max(1, int(GOLD_PERCENT * N))]
silver_idx = idx_sorted[int(GOLD_PERCENT * N):int((GOLD_PERCENT + SILVER_PERCENT) * N)]
train_idx = idx_sorted[int((GOLD_PERCENT + SILVER_PERCENT) * N):]

# Create a 'quality_split' column
df['quality_split'] = 'train' # Default
df.iloc[gold_idx, df.columns.get_loc('quality_split')] = 'gold'
df.iloc[silver_idx, df.columns.get_loc('quality_split')] = 'silver'

# --- Clean Up ---
del semantic_model, fluency_model, fluency_tokenizer
torch.cuda.empty_cache()

print("Evaluation complete.")
```

```python
import numpy as np
from sentence_transformers import SentenceTransformer, util
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
import math
from tqdm import tqdm
import re

# --- Load Models ---
print("Loading models...")
semantic_model = SentenceTransformer('sentence-transformers/LaBSE')
fluency_tokenizer = AutoTokenizer.from_pretrained('aubmindlab/aragpt2-base')
fluency_model = AutoModelForCausalLM.from_pretrained('aubmindlab/aragpt2-base')
fluency_model.eval()

# --- Prepare Text Lists ---
tunisian_texts = df['chunk_text'].astype(str).tolist()
msa_texts_eval = df['cleaned_msa_translation'].astype(str).tolist()

# --- 1. Calculate LaBSE Semantic Similarity ---
print("Calculating LaBSE similarity...")
emb_t = semantic_model.encode(tunisian_texts, convert_to_tensor=True, show_progress_bar=True)
emb_m = semantic_model.encode(msa_texts_eval, convert_to_tensor=True, show_progress_bar=True)
cos_sim = util.cos_sim(emb_t, emb_m).diagonal().cpu().numpy()

# --- 2. Calculate AraGPT2 Perplexity (Fluency) ---
print("Calculating AraGPT2 perplexity...")

def split_sentences_arabic(text):
    """Basic Arabic sentence splitter using punctuation."""
    sentences = re.split(r'(?<=[\.!\؟])\s+', text)
    return [s.strip() for s in sentences if s.strip()]

def calculate_perplexity(text, model, tokenizer, max_length=512):
    sentences = split_sentences_arabic(text)
    losses = []
    was_truncated = False

    for sent in sentences:
        enc = tokenizer(sent, return_tensors="pt", truncation=True, max_length=max_length)
        input_ids = enc.input_ids.to(model.device)

        if input_ids.shape[1] == 0:
            continue  # skip empty segments

        if enc.input_ids.shape[1] >= max_length:
            was_truncated = True

        with torch.no_grad():
            outputs = model(input_ids, labels=input_ids)
            losses.append(outputs.loss.item())

    if not losses:
        return float("inf"), True  # fallback if everything failed

    mean_loss = np.mean(losses)
    ppl = math.exp(mean_loss)
    return ppl, was_truncated

ppls = []
truncated_flags = []

for msa_text in tqdm(msa_texts_eval, desc="Perplexity"):
    ppl, was_truncated = calculate_perplexity(msa_text, fluency_model, fluency_tokenizer)
    ppls.append(ppl)
    truncated_flags.append(was_truncated)

ppls = np.array(ppls)

# --- Debug check (optional) ---
print("Sample perplexities:", ppls[:10])
print("Min perplexity:", np.min(ppls))
print("Max perplexity:", np.max(ppls))

# --- 3. Calculate Composite Score ---
print("Calculating composite score...")

# Normalize LaBSE similarity
cos_norm = (cos_sim - cos_sim.min()) / (cos_sim.max() - cos_sim.min() + 1e-12)

# Clean up PPL values
ppls = np.nan_to_num(ppls, nan=1e6, posinf=1e6, neginf=1e6)

# Log-normalize PPL
log_ppls = np.log(np.clip(ppls, 1e-8, None))  # avoid log(0) or inf
ppl_norm = (log_ppls - log_ppls.min()) / (log_ppls.max() - log_ppls.min() + 1e-12)

# Invert so that lower perplexity = higher score
ppl_score = 1 - ppl_norm

# Composite score
W_SEMANTIC = 0.7
W_FLUENCY = 0.3
composite = W_SEMANTIC * cos_norm + W_FLUENCY * ppl_score

# --- 4. Add scores to DataFrame ---
df['labse_similarity'] = cos_sim
df['aragpt2_perplexity'] = ppls
df['was_truncated'] = truncated_flags
df['norm_labse_similarity'] = cos_norm
df['norm_fluency_score'] = ppl_score
df['composite_score_v2'] = composite

# --- 5. Create Quality Splits ---
print("Creating quality splits...")
idx_sorted = np.argsort(-composite)
N = len(composite)

GOLD_PERCENT = 0.10
SILVER_PERCENT = 0.30

gold_idx = idx_sorted[:max(1, int(GOLD_PERCENT * N))]
silver_idx = idx_sorted[int(GOLD_PERCENT * N):int((GOLD_PERCENT + SILVER_PERCENT) * N)]
train_idx = idx_sorted[int((GOLD_PERCENT + SILVER_PERCENT) * N):]

df['quality_split'] = 'train'
df.iloc[gold_idx, df.columns.get_loc('quality_split')] = 'gold'
df.iloc[silver_idx, df.columns.get_loc('quality_split')] = 'silver'

# --- Clean Up ---
del semantic_model, fluency_model, fluency_tokenizer
torch.cuda.empty_cache()

print("✅ Evaluation complete.")
```
